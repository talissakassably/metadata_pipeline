```python
from Io import Io

io = Io()  # Open class to read data

print('Available sessions:')
for s in io.session_ids:
    print(f'\t{s}')
```

Data is loaded per session, meta data is stored in dataframes

```python
# Load a session
io.load_session('rat-16_181012')

# Meta data is stored in these dataframes
# io.trial_df
# io.unit_df
```

Objects (spike trains/ events etc) can be retreived using the select functions and get_object function

```python
unit_ids = io.select_units(recording_group='BR')  # select units from the Barrel cortex
spiketrains = [io.get_object(uid) for uid in unit_ids]  # laod all spiketrains into a list

trial_ids = io.select_trials(modality='T')  # select all Tactile trials
ev_times = io.get_event_times(trial_ids, 'sample_start')  # return all sample start times
```

The data is stored in a neo object (see the python https://neo.readthedocs.io/en/stable/)

```python
# io.block  # this prints out a whole bunch of metinfo on the structure
# io.block.segments[0]  # each session has only 1 segment
# io.block.segments[0].spiketrains  # spiketrains are stored here
# io.block.segments[0].events  # events are stored here
```

To load LFP data, pass the 'read_lfp' argument to `load_session`

```python
io.load_session('rat-16_181012', read_lfp=True)
lfp_ids = io.select_lfps(recording_group='BR')  # find LFPS in BR
lfp = io.get_object(lfp_ids[0])  # load the first lfp signal
```

Meta data for object (lfps, spiketrains) are stored in their `.annotations` attribute. These are also loaded into the `io.unit_df` pandas dataframe.

```python
unit_annotations = list(spiketrains[0].annotations.keys())
for i in range(5):
    print(unit_annotations[i])
```
