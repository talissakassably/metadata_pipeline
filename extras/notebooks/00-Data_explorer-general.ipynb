# Downloading data 

```python
working_directory = os.getcwd()

newfolder = r'Name of the dataset' 

if not os.path.exists(newfolder):
    os.makedirs(newfolder)
    client_kg = KGClient(host="core.kg.ebrains.eu", token=....get_token())
    dataset = DatasetVersion.from_id("id of the dataset", client=client_kg)
    dataset.show()
    dataset.download(local_path=os.path.join(working_directory, f"{newfolder}") ,client=client_kg, accept_terms_of_use=True)
```

```python
def find_zip_file(dataset_directory): # Find zip file in the folder
    zip_path = glob(os.path.join(dataset_directory, '*.zip'))
    return zip_path[0]
zip_path = find_zip_file(working_directory)
```

```python
def unzip_folder(zip_path, dataset_directory): # Unzip and extract the zipfile in the chosen directory
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(dataset_directory)
        extracted_files = zip_ref.namelist()
        top_level_dirs = [os.path.join(dataset_directory, f.split('/')[0]) for f in extracted_files]
        top_level_dirs = list(set(top_level_dirs))  
    return top_level_dirs[0] if top_level_dirs else dataset_directory

unzip_folder(zip_path, working_directory)
```
